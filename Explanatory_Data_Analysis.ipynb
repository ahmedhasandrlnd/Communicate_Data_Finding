```python
# Monthly usage of the Ford Go Bike system
months=['January','February','March','April','May','June','July','August','September','October','November','December']
graph = sb.catplot(data = df_cleaned_2017, x = 'start_time_month', kind = 'count',order=months);
graph.set_axis_labels('Month', 'Number of Bike Trips'),
graph.fig.suptitle('Ford GoBike Usage by Month in 2017', fontsize = 16, fontweight = 'bold');
graph.set_xticklabels(rotation=90);
```

```python
# Customers vs Subscribers in 2017
customer_2017 = df_cleaned_2017[df_cleaned_2017['user_type'] == "Customer"]['bike_id'].count()
subscriber_2017 = df_cleaned_2017[df_cleaned_2017['user_type'] == "Subscriber"]['bike_id'].count()
customer_prop_2017 = customer_2017 / df_cleaned_2017['bike_id'].count()
subscriber_prop_2017 = subscriber_2017 / df_cleaned_2017['bike_id'].count()
```

```python
# Customers vs Subscribers in 2018
customer_2018 = df_cleaned_2018[df_cleaned_2018['user_type'] == "Customer"]['bike_id'].count()
subscriber_2018 = df_cleaned_2018[df_cleaned_2018['user_type'] == "Subscriber"]['bike_id'].count()
customer_prop_2018 = customer_2018 / df_cleaned_2018['bike_id'].count()
subscriber_prop_2018 = subscriber_2018 / df_cleaned_2018['bike_id'].count()
```

```python
# Customers vs Subscribers in 2019
customer_2019 = df_cleaned_2019[df_cleaned_2019['user_type'] == "Customer"]['bike_id'].count()
subscriber_2019 = df_cleaned_2019[df_cleaned_2019['user_type'] == "Subscriber"]['bike_id'].count()
customer_prop_2019 = customer_2019 / df_cleaned_2019['bike_id'].count()
subscriber_prop_2019 = subscriber_2019 / df_cleaned_2019['bike_id'].count()
```

```python
# Customers vs Subscribers Visuals
# Source: https://matplotlib.org/3.2.1/gallery/pie_and_polar_charts/pie_demo2.html#sphx-glr-gallery-pie-and-polar-charts-pie-demo2-py
labels = 'Customer', 'Subscriber'

plt.figure(figsize = [8, 5])
prop_2017 = [customer_prop_2017, subscriber_prop_2017]
plt.pie(prop_2017, labels=labels, autopct='%1.1f%%', shadow=True)
plt.suptitle('Ford GoBike System - Customers vs. Subscribers in 2017', y=1.05, fontsize=16, fontweight='bold');
```

```python
# Customers vs Subscribers Visuals
# Source: https://matplotlib.org/3.2.1/gallery/pie_and_polar_charts/pie_demo2.html#sphx-glr-gallery-pie-and-polar-charts-pie-demo2-py
plt.figure(figsize = [8, 5])
prop_2018 = [customer_prop_2018, subscriber_prop_2018]
plt.pie(prop_2018, labels=labels, autopct='%1.1f%%', shadow=True)
plt.suptitle('Ford GoBike System - Customers vs. Subscribers in 2018', y=1.05, fontsize=16, fontweight='bold');
```

```python
# Customers vs Subscribers Visuals
# Source: https://matplotlib.org/3.2.1/gallery/pie_and_polar_charts/pie_demo2.html#sphx-glr-gallery-pie-and-polar-charts-pie-demo2-py
plt.figure(figsize = [8, 5])
prop_2019 = [customer_prop_2019, subscriber_prop_2019]
plt.pie(prop_2019, labels=labels, autopct='%1.1f%%', shadow=True)
plt.suptitle('Ford GoBike System - Customers vs. Subscribers in 2019', y=1.05, fontsize=16, fontweight='bold');
```

#### Observation: The subscriber proportion increased from 2017 to 2018, but it decreased from 2018 to 2019, which suggests that some subscriber withdrew their membership.

```python
# source: https://github.com/chelseymarie6/Communicate-Data-Findings/blob/master/Communicate_Data_Slide_Deck.ipynb
# Weekday order
weekday = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
df_cleaned_2017['start_time_weekday'] = pd.Categorical(df_cleaned_2017['start_time_weekday'], categories= weekday, ordered=True);

plt.figure(figsize=[11, 8]);

# Heatmap - Customers
plt.subplot(1, 2, 1)
df_customer = df_cleaned_2017.query('user_type == "Customer"').groupby(['start_time_hour', 'start_time_weekday'])['bike_id'].size().reset_index();
df_customer = df_customer.pivot('start_time_hour', 'start_time_weekday', 'bike_id')
sb.heatmap(df_customer)

plt.title('Customer', y=1.05)
plt.xlabel('Weekday', labelpad = 16)
plt.ylabel('Start Time Hour',  labelpad = 16)

# Heatmap - Subscribers
plt.subplot(1, 2, 2)
df_subscriber = df_cleaned_2017.query('user_type == "Subscriber"').groupby(['start_time_hour', 'start_time_weekday'])['bike_id'].size().reset_index();
df_subscriber = df_subscriber.pivot('start_time_hour', 'start_time_weekday', 'bike_id')
sb.heatmap(df_subscriber)

plt.title("Subscriber", y=1.05)
plt.xlabel('Weekday', labelpad = 16)
plt.ylabel('Start Time Hour', labelpad = 16);

plt.suptitle('Ford GoBike System Rush Hours in 2017', fontsize = 16, fontweight = 'bold');
```

### Summary:
There are two types of clients using the system: Subscribers and Customers. Subscribers are primarily daily commuters, having short trips to and from work, who rent a bike on weekdays at 8-9am and 5-6pm. Customers are usually tourists or occasional riders who use the system mainly on weekends to explore the Bay Area. In 2019, some commuter subscribers seem to withdraw their membership and tend to use the bike service as non-member customers as indicated in customer vs subscriber proportion and rush hour patterns by user type and start day graphs.
